# Model for the spread of COVID-19

```python
import matplotlib.pyplot as plt
S, I, R= [], [], []
```

```python
beta = 2.5 / 14 #infection rate: 2.5 other people in 14 days
gamma = 1/14 #it takes 14 days to recover
So = 17999999 #18 million people -1 infected
Io = 1 #1 infected
Ro = 0 #recoverd
N = So + Io + Ro #total number of people
```

```python
for j in range(500):
    new_infections = beta * Io * So / N
    new_recoveries = gamma * Io
    Sn = So - new_infections
    Rn = Ro + new_recoveries
    In = Io + new_infections - new_recoveries
    S.append(Sn)
    I.append(In)
    R.append(Rn)
    So = Sn
    Io = In
    Ro = Rn
```

```python
plt.plot(S)
plt.plot(I)
plt.plot(R)
plt.legend(["Susceptible", "Infected", "Recoverd"], loc ="lower right")
plt.xlabel("days")
plt.ylabel("number of people")
```
