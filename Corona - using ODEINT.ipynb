# Model for the spread of COVID-19
## (SIR model solved with ODE)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# defining our ODE system of equations
def SIRmodel(z, t, beta, gamma):
    S, I, R = z
    dSdt = -beta * S* I / N
    dIdt =  beta * S * I / N - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]
```

```python
# defining the initial conditions
S0, I0, R0 = 17999999, 1, 0  #18 million people, 1 infected, 0 recovered
beta = 2.5 / 14              #infection rate: 1 person infects 2.5 other people in 14 days
gamma = 1/14                 #it takes 14 days to recover
N = S0 + I0 + R0             #total number of people
number_of_days = 500         #number of days for the simulation to run
```

```python
# solving the ODE
time = np.arange(0, number_of_days)
Z = odeint(SIRmodel, [S0, I0, R0], time, args=(beta, gamma))
S = Z[:,0]
I = Z[:,1]
R = Z[:,2]
```

```python
plt.plot(time,S, label = "Susceptible")
plt.plot(time,I, label = "Infected")
plt.plot(time,R, label = "Recovered")
plt.grid()
plt.xlabel("days")
plt.ylabel("number of people")
plt.legend()
```
