# Model for the spread of COVID-19
## (SIR model extended; you can get ill again)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_days = 1000
t = np.arange(0,number_of_days)
S = np.zeros(number_of_days)
I = np.zeros(number_of_days)
R = np.zeros(number_of_days)
```

```python
S[0], I[0], R[0] = 17999999, 1, 0  # 18 million people, 1 infected, 0 recovered
beta = 2.5 / 14                    #infection rate: 2.5 other people in 14 days
gamma = 1/14                       #it takes 14 days to recover
alfa = 1/150                       #after 150 days you can get ill again
N = S[0] + I[0] + R[0]             #total number of people
```

```python
for d in range(number_of_days-1):
    new_infections = beta * I[d] * S[d] / N
    new_recoveries = gamma * I[d]
    new_susceptibles = alfa * R[d]
    S[d+1] = S[d] - new_infections + new_susceptibles
    R[d+1] = R[d] + new_recoveries - new_susceptibles
    I[d+1] = I[d] + new_infections - new_recoveries    
```

```python
plt.plot(t,S, label = "Susceptible")
plt.plot(t,I, label = "Infected")
plt.plot(t,R, label = "Recovered")
plt.xlabel("days")
plt.ylabel("number of people")
plt.legend()
```
